## Categorizing the various factors present in the dataset into 3 category viz: Job related, individual related and psychoological and socio-cultural relatyed factors

1. Job-Related Factors

'Environment Satisfaction': This relates to the physical and social work environment.

'Job Satisfaction': This is directly related to the employee's job.

'Business Travel': This is a requirement of some jobs.

'Department': This is the specific area of the company where the employee works.

'JobLevel': This relates to the employee's position within the company.

'JobRole': This is the specific role or duties of the employee.

'Monthly Income': This is the compensation the employee receives for their work.

'Percent Salary Hike': This is related to the employee's compensation.

'Standard Hours': This relates to the employee's work schedule.

'Stock Option Level': This is a form of compensation.

'Training Times Last Year': This relates to the opportunities for development and growth in the job.

'Years At Company': This relates to the duration of the employee's job.

'Years Since Last Promotion': This relates to career advancement within the job.

'Years With Current Manager': This relates to the employee's relationship with their supervisor.

'Manager Job Involvement': This relates to the employee's relationship with their supervisor.

'Manager Performance Rating': This relates to the employee's relationship with their supervisor.

2. Individual-Related Factors

'Age': This is a personal demographic factor.

'Education': This is the employee's personal level of formal education.

'Education Field': This is the field in which the employee received their education.

'Gender': This is a personal demographic factor.

'Marital Status': This is a personal demographic factor.

'Num Companies Worked': This relates to the employee's personal work history.

'Over 18': This is a personal demographic factor.

'Total Working Years': This relates to the employee's personal work history.

3. Psychological and Socio-Cultural Factors

'Work-Life Balance': This relates to the employee's ability to balance their work and personal life.

'Distance From Home': This could relate to the stress of commuting or being away from family.

'Employee Count': While not directly a psychological or socio-cultural factor, this could influence the social dynamics at the workplace.

### Correcting data Imbalance 

```python
"""Visualizing the original distribution of data"""
# Percentage distribution of the dataset
df['label'].value_counts().plot.pie(explode=[0.1,0.1],autopct='%1.1f%%',
                                   shadow=True, wedgeprops={'alpha':0.2})
```

Due to the high level of imbalance the dataset will be resample using the under sample method and also try the effect of oversampling method due to the small size of the data

```python
""" Before resampling the dataset needs to be separated into dependend and independent variables"""
# dependent or target variable
y=df['label']
# Independent or predictor variable
X=df.drop(['label', 'Left_Prob'], axis=1)
```

```python
X.columns
```

## Dropping irrelevant columes or columns with less impact on employees Attriion 

```python
X.drop(['EmployeeID','EducationField','StockOptionLevel',
        'TrainingTimesLastYear'], inplace=True, axis=1)
```

### OverSampling 

```python
! pip install imblearn
```

```python
"""Resampling the Data"""
from imblearn.over_sampling import SMOTE

# Assuming you have features X and target variable y
smote = SMOTE(random_state=42)
Oversampled_X, Oversampled_y = smote.fit_resample(X, y)
print('Before resampling',X.shape)
print(y.value_counts())
print('=====================================================')
print('After resampling',Oversampled_X.shape)
print(Oversampled_y.value_counts())
```

### Undersampling

```python
""" Data Resampling"""
from imblearn.under_sampling import RandomUnderSampler

# Initialize the RandomUnderSampler
under_sampler = RandomUnderSampler(random_state=50)

# Perform under-sampling
resampled_X, resampled_y = under_sampler.fit_resample(X, y)

print('Before resampling',X.shape)
print(y.value_counts())
print('=====================================================')
print('After resampling',resampled_X.shape)
print(resampled_y.value_counts())
```

```python
"""Visualizing the distribution of the resampled data"""
# Percentage distribution of the dataset
Oversampled_y.value_counts().plot.pie(explode=[0.1,0.1],autopct='%1.1f%%',
                                   shadow=True, wedgeprops={'alpha':0.2})
```

### splitting data into a training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Oversampled_X,Oversampled_y, test_size=0.2, random_state=100)
print('X_train :',X_train.shape)
print('X_test :',X_test.shape)
```

## Normalization 
Scaling the the predictor variable  so as to correct the little outlier found 

```python
from sklearn.preprocessing import StandardScaler
# Step 1: Fit the StandardScaler on the training set
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

# Step 2: Use the same scaling parameters on the test set
X_test = scaler.transform(X_test)
```

```python
X_train
```

### Model seclection and training with some hyperparameter turning
The following models were models of choice

### Traing and Testing a Logistic Regression Algorithm with hyperparameter turning (model optimization)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Defining the parameter grid for hyperparameter tuning
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Inverse of regularization strength
    'penalty': ['l1', 'l2'],  # Regularization penalty ('l1' for L1 regularization, 'l2' for L2 regularization)
    'solver': ['liblinear']  # Optimization algorithm
}
"""instantiating the model with"""
# Create an instance of the Logistic Regression classifier with wei
log = LogisticRegression()

# Create an instance of GridSearchCV for hyperparameter tuning
LR_grid_search = GridSearchCV(estimator=log, param_grid=param_grid, cv=5)

# Fit the GridSearchCV on the training data
LR_grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the best estimator
best_params = LR_grid_search.best_params_
best_estimator = LR_grid_search.best_estimator_
print('best parameter', best_params)

# Evaluate the model on the test set
y_pred = LR_grid_search.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)*100
print("Accuracy:", accuracy)
```

```python
print(y_pred)
print(y_test)
```

### Evaluating Logistic Regression model

```python
from sklearn.metrics import confusion_matrix, classification_report
print('Training Accuracy :',LR_grid_search.score(X_train, y_train))
print('Testing Accuracy :',LR_grid_search.score(X_test,y_test))
con_mat = pd.DataFrame(confusion_matrix(LR_grid_search.predict(X_test), y_test),
            columns = ['Predicted: Stayed', 'Predicted:Left'],
            index = ['Actual:Stayed', 'Actual:Left'])
print('=================================================================')

print(con_mat)
print('\nCLASSIFICATION REPORT\n')
print(classification_report(LR_grid_search.predict(X_test), y_test,
                            target_names =['Stayed','Left']))
print('=================================================================')

def plot_confusion_matrix(cm, class_names):
    """
  Returns a matplotlib figure containing the plotted confusion matrix.

  Args:
    cm (array, shape = [n, n]): a confusion matrix of integer classes
    class_names (array, shape = [n]): String names of the integer classes
  """
    import itertools
    # Normalize the confusion matrix.
    cm = np.around(cm.astype("float") / cm.sum(axis=1)[:, np.newaxis], decimals=2)

    figure = plt.figure(figsize=(8, 8))
    plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
    plt.title("Confusion matrix")
    plt.colorbar()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45)
    plt.yticks(tick_marks, class_names)

    # Use white text if squares are dark; otherwise black.
    threshold = cm.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        color = "white" if cm[i, j] > threshold else "black"
        plt.text(j, i, cm[i, j], horizontalalignment="center", color=color)

    plt.tight_layout()
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.show()
    
    # Conusion metrix
cm = confusion_matrix(y_test, LR_grid_search.predict(X_test)) #confusion matrix
classname = ['Stayed','Left']
plot_confusion_matrix(cm, classname) #plot confusion matrix
```

### Traing and Testing a Random Forest Algorithm with hyperparameter turning (model optimization)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV

param_grid = {
    'n_estimators': [100,200],
    'max_features': ['sqrt', 'log2'],
    'criterion' :['gini', 'entropy']
}


rf_grid = HalvingGridSearchCV(RandomForestClassifier(random_state=0), param_grid, cv=10)

# fitting the model for grid search
rf_grid.fit(X_train, y_train)
# print best parameter after tuning
print(rf_grid.best_params_)
y_pred = rf_grid.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)*100
print("Accuracy:", accuracy)
```

## Training and Hyperturning an XGB model

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score



# Define the XGB model
model = xgb.XGBClassifier()

# Define the hyperparameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# Create GridSearchCV object
xgb_grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=3, verbose=1)

# Fit the model
xgb_grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = xgb_grid_search.best_params_
print("Best Parameters:", best_params)

# Make predictions on the test set using the best model
y_pred = xgb_grid_search.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
```
